```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression

employee_data = pd.read_csv('Salary_Data.csv')
print(employee_data)

experience_years = employee_data["YearsExperience"]
salary = employee_data["Salary"]

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(experience_years, salary, test_size=0.33, random_state=0)

# Create a linear regression model and fit it to the training data
salary_regressor = LinearRegression()
salary_model = salary_regressor.fit(X_train.values.reshape(-1, 1), Y_train.values.reshape(-1, 1))

# Print the coefficients and intercept of the linear model
print("Coefficients:", salary_model.coef_)
print("Intercept:", salary_model.intercept_)

# Predict the target variable on the test set
salary_predictions = salary_model.predict(X_test.values.reshape(-1, 1))

# Plot the real data and the linear model
plt.title("Salary/Years of Experience")
plt.ylabel("Salary $")
plt.xlabel("Years of Experience")
plt.scatter(X_test, Y_test, color="blue", label="Real Data")
plt.plot(X_test, salary_predictions, color="red", label="Linear Model")
plt.legend()
plt.show()

# Calculate and print the mean squared error of the model
mse = mean_squared_error(Y_test, salary_predictions)
print("Mean Squared Error:", mse)
```
